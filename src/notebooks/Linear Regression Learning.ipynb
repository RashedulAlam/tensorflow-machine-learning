```python
from tensorflow.keras.utils import plot_model
plot_model(model, show_shapes=True)
```

```python
y_pred = model.predict(tf.expand_dims(X_test, axis=1));
y_pred
```

```python
def plot_predictions(train_data=X_train,
                    train_labels=y_train,
                    test_data=X_test,
                    test_labels=y_test,
                    predictions=y_pred):
  """
  Plots traning the data, test data and compares predictions
  """
  plt.figure(figsize=(10, 7))
  # Plot training data in blue
  plt.scatter(train_data, train_labels, c="b", label="Training data")
  # Plot testing data in green
  plt.scatter(test_data, test_labels, c="g", label="Testing data")
  # Plot the predictions in red (predictions were made on the test data)
  plt.scatter(test_data, predictions, c="r", label="Predictions")
  # Show the legend
  plt.legend();

  plt.show();

plot_predictions(X_train, y_train, X_test, y_test, y_pred);
```

## Experiments


```python
tf.random.set_seed(42);

model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1,),
    ]);


model_1.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

model_1.fit(tf.expand_dims(X_train, axis=1), y_train, epochs=100, verbose=0)

model_1.summary()
```

```python
y_pred_1 = model_1.predict(tf.expand_dims(X_test, axis=1));
y_pred_1

plot_predictions(X_train, y_train, X_test, y_test, y_pred_1)
```

```python
def model_evalution_matries(y_test, y_pred):
  mae = tf.keras.losses.mae(y_test, y_pred.squeeze());
  mse = tf.keras.losses.mean_squared_error(y_test, y_pred.squeeze());

  print(f"MAE: {mae}\nMSE: {mse}");

  return mae, mse;

model_evalution_matries(y_test, y_pred_1)
```

```python
tf.random.set_seed(42);

model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1),
]);

model_2.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mse']);

model_2.fit(tf.expand_dims(X_train, axis=1), y_train, epochs=10, verbose=0);

model_2.summary();
```

```python
y_pred_2 = model_2.predict(tf.expand_dims(X_test, axis=1));
plot_predictions(X_train, y_train, X_test, y_test, y_pred_2)
model_evalution_matries(y_test, y_pred_2)
```

```python
tf.random.set_seed(42);


model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1),]);


model_3.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae']);


model_3.fit(tf.expand_dims(X_train, axis=1), y_train, epochs=500, verbose=0);

model_3.summary();
```

```python
y_pred_3 = model_3.predict(tf.expand_dims(X_test, axis=1));

plot_predictions(X_train, y_train, X_test, y_test,y_pred_3);

model_evalution_matries(y_test, y_pred_3)

y_pred_3
```

```python
model_results  = [["model_1", model_evalution_matries(y_test, y_pred_1)],
                  ["model_2", model_evalution_matries(y_test, y_pred_2)],
                  ["model_3", model_evalution_matries(y_test, y_pred_3)]];


model_results
```
